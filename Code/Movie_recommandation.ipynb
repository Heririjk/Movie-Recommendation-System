```python
import pandas as pd

# Lees de csv-bestanden in pandas dataframes
movies = pd.read_csv('movies_metadata.csv')
ratings = pd.read_csv('ratings.csv')
credits = pd.read_csv('credits.csv')

# Zet de kolom 'id' in movies_metadata.csv en 'movieId' in ratings.csv om naar hetzelfde datatype als 'id' in movies_metadata.csv
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')
ratings['movieId'] = pd.to_numeric(ratings['movieId'], errors='coerce')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')

# Hernoem de kolom 'movieId' in 'id' in het ratings dataframe
ratings = ratings.rename(columns={'movieId': 'id'})

# Merge de drie dataframes op basis van de kolom 'id'
merged = pd.merge(pd.merge(movies, credits, on='id', how='left'), ratings, on='id', how='left')

# Gebruik een lambda functie om enkel de namen van de acteurs weer te geven
#merged['cast'] = merged['cast'].apply(lambda x: [i['name'] for i in eval(x)] if isinstance(x, str) else [])

# Nu heb je een dataframe met de metadata van alle films, de bijbehorende credits en ratings indien beschikbaar, en enkel de namen van de acteurs in de "cast" kolom
```

```python
# Verwijder de kolommen zonder ratings uit de oorspronkelijke DataFrames
movies = movies[['id', 'title', 'release_date', 'genres']].dropna()
ratings = ratings[['userId', 'id', 'rating']].dropna()

# Maak de 'title' kolom uniek
movies = movies.drop_duplicates(subset=['title'])

# Verwijder de duplicaten in de 'merged' DataFrame op basis van de 'id' kolom en 'title' kolom
merged = merged.drop_duplicates(subset=['id', 'title'])

# Verwijder eventuele rijen met NaN-waarden
merged = merged.dropna()
```

```python
merged["genres"] = merged["genres"].apply(lambda x: [genre["name"].split()[0] for genre in eval(x)])

```

```python
merged["cast"] = merged["cast"].apply(lambda x: [cast["name"].split()[0] for cast in eval(x)])
```

```python
merged["production_countries"] = merged["production_countries"].apply(lambda x: [production_countries["name"].split()[0] for production_countries in eval(x)])
```

```python
final = merged[["overview", "original_title", "id", "original_language", "genres"]]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv=TfidfVectorizer(min_df=3, max_features=None, strip_accents= "unicode", analyzer='word', token_pattern=r'\w{1,}', ngram_range=(1,3), stop_words='english')
```

```python
tfv_matrix=tfv.fit_transform(final["overview"])
```

```python
index1=pd.Series(final.index, index=final["original_title"]).drop_duplicates()
```

```python
index1
```

```python
def give_recommendation(title,sig=sig):
    idx=index1[title]
    sigmoid_score= list(enumerate(sig[idx]))
    sigmoid_score=sorted(sigmoid_score, key=lambda x:x[1],reverse=True)
    sigmoid_score=sigmoid_score[1:6]
    movie_indices=[i[0] for i in sigmoid_score]
    return final["original_title"].iloc[movie_indices]
```

```python
give_recommendation("Blue in the Face")
```
